```python
import numpy as np
import pandas as pd
from datetime import datetime
```

```python
dataset = pd.read_excel(r'C:\Users\Ivan Sayapin\Desktop\S&P 500\S&P_500_Data.xlsx')
```

```python
dataset
```

```python
X = dataset.iloc[:, [0, 1]].values
df = pd.DataFrame(data=X)
```

```python
i = 0
while i < len(df):
    old_date = df.iat[i,0]
    datetime_object = datetime.strptime(old_date,'%b %d, %Y')
    new_date = datetime_object.strftime("%d/%m/%y")
    df = df.replace(old_date,new_date)
    i = i + 1
```

```python
#example dataset with tweets and date
Y = ['14/02/18','02/02/18', '19/01/18']
example = pd.DataFrame(data=Y)
example.columns = ['date']
```

```python
example
```

```python
#search for price that day
prices = pd.DataFrame(data = [])
i = 0
j = 0

while i < len(example):
    while j < len(df):
        if example.iat[i,0] == df.iat[j,0]:
            prices = prices.append(pd.DataFrame(data = [df.iat[j,1]]))
            i = i + 1
            break
        j = j + 1
```

```python
prices = prices.reset_index(drop=True)       
prices.columns = ['S&P 500']

final = pd.concat([example,prices], axis = 1)
final.columns = ['date','S&P 500 price']
```
