### Get Tweets

```python
#Pull tweets given screen name; max 200 tweets
def get_tweets(screen_name):
    api = tweepy.API(auth)
    alltweets = []
    new_tweets = api.user_timeline(screen_name=screen_name,count=200, include_rts = True, tweet_mode='extended')
    alltweets.extend(new_tweets)
    outtweets = [[tweet.id_str, 
                  tweet.created_at, 
                  tweet.full_text.encode("utf-8"), 
                  tweet.source,
                  tweet._json['retweet_count'],
                  tweet._json['favorite_count'],
                  tweet._json['user']['location'],
                  tweet._json['user']['followers_count']] for tweet in alltweets]
    return outtweets
    print (len(alltweets))
```

```python
def get_all_tweets(targets):
    tweets = []
    for target in targets:
        tweets = tweets + get_tweets(target)
    return tweets
```

### Vader Sentiment

```python
analyser = SentimentIntensityAnalyzer()
def get_sentiment_scores(sentence):
    snt = analyser.polarity_scores(sentence)
    return(snt)
```

### NLTK pos/neg words sentiment

```python
from nltk.corpus import opinion_lexicon
from nltk.stem import PorterStemmer
ps = PorterStemmer()
```

```python
positive_words =  opinion_lexicon.words('positive-words.txt')
negative_words =  opinion_lexicon.words('negative-words.txt')          
```

```python
sentcols = ['pos','neg', 'wordcount']
sentiment = []
sentDF = pd.DataFrame(sentiment, columns=sentcols)
for tweet in range(0,len(twDF)):
    words = word_tokenize(str(twDF['Text'][tweet]))
    pos_count = 0
    neg_count = 0
    word_count = len(words)
    for i in range(1,len(words)):
        if ps.stem(words[i]) in positive_words:
            pos_count+=1
        elif ps.stem(words[i]) in negative_words:
            neg_count+=1
    sentDF.loc[tweet] = [pos_count, neg_count, word_count]
print(sentDF)
```

```python
twsentDF = pd.concat([twDF,sentDF], axis=1)
twsentDF
```

### Data Exploration/ Fun Graphs

```python
net_pos = twsentDF["pos"]-twsentDF["neg"]
plt.bar(twsentDF["date"][net_pos>=0],net_pos[net_pos>=0], width = 0.1, color = 'b') 
plt.bar(twsentDF["date"][net_pos<0],net_pos[net_pos<0], width = 0.1, color = 'r') 
plt.plot(twsentDF["date"],[net_pos.mean()]*len(twsentDF), color = '#000000') 
plt.title('Net Positivity')
plt.show()
print("Mean Net Positivity: ", net_pos.mean())
```

```python
wnet_pos = (twsentDF["pos"]-twsentDF["neg"])*twsentDF["Retweets"]/twsentDF["Retweets"].mean()
plt.bar(twsentDF["date"][wnet_pos>=0],wnet_pos[wnet_pos>=0], width = 0.1, color = 'b') 
plt.bar(twsentDF["date"][wnet_pos<0],wnet_pos[wnet_pos<0], width = 0.1, color = 'r') 
plt.plot(twsentDF["date"],[wnet_pos.mean()]*len(twsentDF), color = '#000000') 
plt.title('Weighted Net Positivity')
plt.show()
print("Mean Net Positivity (Weighted by Retweets): ", wnet_pos.mean())
```
